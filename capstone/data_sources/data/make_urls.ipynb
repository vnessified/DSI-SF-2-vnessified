```python
import glob
```

```python
import os
```

```python
rest_ids = [x for x in glob.glob('./*') if os.path.isdir(x)]
```

```python
rest_ids
```

```python
rest_id_food_ids = []
for ri in rest_ids:
    rest_id_food_ids.extend([x for x in glob.glob(ri+'/*') if os.path.isdir(x)])
```

```python
rest_id_food_ids
```

```python
# http://pfid.rit.albany.edu/viewer.php?rest_id=3&food_id=28&type=2
```

```python
urls_to_scrape = [
    'http://pfid.rit.albany.edu/viewer.php?rest_id='+x.split('/')[1]+'&food_id='+x.split('/')[2]+'&type=2'
    for x in rest_id_food_ids
]
```

```python
urls_to_scrape
```

```python
url_dict = {}

for rif, url in zip(rest_id_food_ids, urls_to_scrape):
    rid = rif.split('/')[1]
    fid = rif.split('/')[2]
    if rid in url_dict.keys():
        url_dict[rid][fid] = url
    else:
        url_dict[rid] = {}
        url_dict[rid][fid] = url
```
